```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("dados.csv")
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df)
df2 = df1.iloc[:,1:]
df2.head()
```

```python
frequent_items = apriori(df2, min_support=0.05, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
type(rules)
rules.shape
```

```python
rules
```

```python
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.6) & (rules['support'] >= 0.2)]
```

```python
csv_file = "rules.csv"
rules_df = rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.6) & (rules['support'] >= 0.2)]
rules_df.to_csv(csv_file, index=False)
```
